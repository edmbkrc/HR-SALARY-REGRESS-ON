```python
miss=hr.isnull().sum()
miss[miss>0]
```

```python
data=hr.copy()#kopyası üzerinden çalışmaya başladık
del data["Attrition"]
```

```python
data.isnull().sum()
```

```python
missing=data[data["MonthlyIncome"].isnull()]
filled=data[data["MonthlyIncome"].notnull()]
```

```python
missing.shape, filled.shape
```

```python
#Simple Linear
```

```python
filled.corr()
```

```python
abs(filled.corr()["MonthlyIncome"]).sort_values(ascending=False)
```

```python
from scipy import stats
est=stats.linregress(x=filled["TotalWorkingYears"],y=filled["MonthlyIncome"])
est
```

```python
est.slope
```

```python
est.intercept
```

```python
def predict(x):
    return est.slope * x + est.intercept
```

```python
predict(2)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(filled["MonthlyIncome"],predict(filled["TotalWorkingYears"]))
```

```python
mean_squared_error(filled["MonthlyIncome"],filled["TotalWorkingYears"].apply(predict))**0.5
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(filled["MonthlyIncome"],filled["TotalWorkingYears"])
plt.plot(filled.TotalWorkingYears,predict(filled.TotalWorkingYears),c="r")
```
